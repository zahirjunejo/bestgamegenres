```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
appstore_games = pd.read_csv('appstore_games.csv')
appstore_games.head()
```

```python
appstore_games = appstore_games.fillna(0)
appstore_games
```

```python
# We need to first filter out the games that got a higher user rating count and higher average rating at the same time.
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
ax = sns.scatterplot(x="Average User Rating", y="User Rating Count", palette=cmap, data=appstore_games)
```

```python
# Lets see whats the genres of the games
appstore_games = appstore_games[appstore_games["Average User Rating"] > 4]
appstore_games.head()
```

```python
# And the genres worth going for are....
appstore_games.groupby("Genres").count().sort_values(by="ID", ascending=False)[["ID"]].rename(columns={"ID": "# of games"})
```
